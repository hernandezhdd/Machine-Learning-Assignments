```
import pandas as pd

df = pd.read_json('sample_data/anscombe.json') #Leo los datos de anscombe
incognitas = {
'x': [16, 8, 16, 16],
'y': [8, 6, 10, 6],
}
numeros = df.Series.unique() #Nombres de los datasets
```

```
import matplotlib.pyplot as plt

for numero in numeros:          #Hago un gráfico para cada dataset
  plt.figure(figsize=(14,6))
  plt.scatter(df.loc[df.Series == numero]['X'], df.loc[df.Series == numero]['Y'])
  plt.scatter(incognitas['x'], incognitas['y'])
  plt.title('Dataset '+numero)
  plt.show()
```
